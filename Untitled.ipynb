# Implementation of Logistic Regression

```python
def segmoid(x):
    return 1.0/(1.0+np.exp(-x))


def hypothesis(X,theta):
    
    return segmoid(np.dot(X,theta))

def error(X,y,theta):
    
    hi= hypothesis(X,theta)
    
    e= -1*np.mean((y*np.log(hi) + ((1-y)*np.log(1-hi))))
    
    return e

def gradient(X,y,theta):
    hi = hypothesis(X,theta)
    grad = -np.dot(X.T,(y-hi))
    
    m = X.shape[0]
    return grad/m

def gradient_descent(X,y,lr=0.8,max_itr=500):
    n = X.shape[1]
    theta = np.zeros((n,1))
    
    error_list = []
    
    current_error=0
    for i in range(max_itr):
        
        err = error(X,y,theta)
        error_list.append(err)
        
        
    
        
        grad = gradient(X,y,theta)
        theta = theta - lr*grad
        
        
        
    return (theta,error_list)



```

```python
ones = np.ones((x_train.shape[0],1))
X_New_train = np.hstack((ones,x_train))

print(X_New_train.shape)
print(X_New_train)
y_train = y_train.reshape((-1,1))
```

```python
theta,error_list = gradient_descent(X_New_train,y_train)
```

```python
#test data

x_test = pd.read_csv('Logistic_X_Test.csv')
xt_mean = x_test.mean(axis=0)
xt_std = x_test.std(axis=0)
x_test = (x_test-xt_mean)/xt_std
x_test = np.array(x_test)
ones = np.ones((x_test.shape[0],1))
X_New_test = np.hstack((ones,x_test))


y_train = y_train.reshape((-1,1))

X_New_test
```

```python
#making predictions

def predict(X,theta):
    
    h = hypothesis(X,theta)
    
    output  = np.zeros(h.shape)
    
    output[h>=0.5] = 1
    output = output.astype('int')
    
    return output


XT_preds=predict(X_New_train,theta)
Xt_preds = predict(X_New_test,theta)
Xt_preds
```

```python
def accuracy(actual,preds):
    actual = actual.astype('int')
    actual = actual.reshape((-1,1))
    
    acc = np.sum(actual==preds)/actual.shape[0]
    
    return acc*100
    
```
